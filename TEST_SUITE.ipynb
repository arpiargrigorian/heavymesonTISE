## TEST SUITE NOTEBOOK

This notebook runs a series of tests for a variety of energy levels. 

**Solutions** just demonstrates that the zeta values fall within an interval shown in Eitchen+1978. The values they find are only plotted, not in a table, so I can only guess what they are and hence that large value of rtol. The only exception to this is the solution for l = 0 and lambda = 0, which is the Airy solution and known exactly. 

```python
from find_best_zeta_library import *
import numpy.testing as npt
```

```python
## Init Potential Numbers
el = 0
coulomb_constant = 1.3775510204081631
quantum_n = 1


# Init Rho
rho = np.linspace(0, 30, num = 1000) # The Full array of rho

rho_min_cut = 5 
rho_max_cut = 250

rho_cut_interval = (rho_min_cut, rho_max_cut)


# Init Zeta

zeta_init_interval = (.5, 2.5)
zeta_num_array = [80, 40, 40, 40]

simulation_results = iter_find_zeta(el, quantum_n, coulomb_constant, rho, rho_cut_interval, zeta_init_interval, zeta_num_array)

## Look at the Solution to make sure all is well.

plt.plot(simulation_results[2], simulation_results[4])
plt.plot(simulation_results[3], simulation_results[5], linestyle = '--')

npt.assert_allclose(simulation_results[0][-1], 1, rtol = .1)
```
